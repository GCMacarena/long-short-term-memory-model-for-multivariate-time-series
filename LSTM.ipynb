```python
# Resumen estadístico de las columnas numéricas del DataFrame
df.describe()
```

```python
# Verifico si hay nulos
df.isnull().sum()
```

```python
# Transformo la columna 'date' para posteriormente poder normalizarla

df["date"] = pd.to_datetime(df["date"], format="%Y/%m/%d") #Esta línea convierte la columna "date" del DataFrame en objetos de tipo datetime utilizando la función pd.to_datetime()
df = df.set_index("date") #Esta línea establece la columna "date" como el índice de las filas del DataFrame
df.head()
```

```python
# La columna a predecir será greatest losses direction porque proporciona información crítica sobre la dirección en la que se están experimentando las mayores pérdidas en el conflicto.
# Siendo que sus valores son nombres de ciudades, asigno un número único a cada nombre (que posteriormente podré identificar)
# Pero antes de todo esto, le asigno valor 0 a NaN para evitar futuros errores

df['greatest losses direction'].fillna(0, inplace=True) # NaN = 0

# Diccionario para asignar números únicos a cada nombre
name_to_number = {}
number_to_name = {}
next_number = 0

for index, row in df.iterrows():  #Recorro la columna "greatest losses direction" y asigno números únicos a cada nombre
    name = row['greatest losses direction']
    if name not in name_to_number:
        name_to_number[name] = next_number
        number_to_name[next_number] = name
        next_number += 1

    df.at[index, 'greatest losses direction'] = name_to_number[name] # Reemplazo el nombre por el número en la columna
```

```python
# Guardo los datos actualizados en un nuevo archivo CSV
df.to_csv('russia_losses_equipment_2.csv', index=False)
```

```python
# Para identificar los nombres en base a los números, puedo usar el diccionario number_to_name anteriormente creado
number = 3  # Ejemplo de número a identificar
name = number_to_name.get(number)
if name:
    print(f'El nombre correspondiente al número {number} es {name}')
else:
    print(f'No se encontró ningún nombre correspondiente al número {number}')
```

```python
# Normalizo los valores de cada columna restando el valor mínimo de la columna y luego
# dividiendo por la diferencia entre el valor máximo y el valor mínimo de la columna.
# El resultado es una columna con valores normalizados en el rango de 0 a 1.
# Además, reemplazo los valores nulos anteriormente señalados con valor 0

for c in df.columns:
    if df[c].dtype != 'float': # Señalo los float para posteriormente transformarlos en números enteros
        df[c] = pd.to_numeric(df[c], errors='coerce')  # pd.to_numeric() para convertir los valores de la columna en numeros no decimales (int) | errors='coerce' para convertir los valores no numéricos en NaN
    df[c] = (df[c] - df[c].min()) / (df[c].max() - df[c].min())

df.fillna(0, inplace=True)
```

# **LSTM Series temporales multivariadas (1)**

```python
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# Visualizo todas las columnas

df.plot.line(figsize=(20, 5))
plt.show()
```

```python
# Divido el DataFrame en conjuntos de entrenamiento y de prueba
df_train = df[df.index < '2023-01-01'] # Datos de entrenamiento que contiene todas las filas del DataFrame cuyas fechas son anteriores al 1 de enero de 2023
df_test = df[df.index >= '2023-01-01'] # Datos de prueba que contiene todas las filas del DataFrame cuyas fechas igual o posterior al 1 de enero de 2023
df_train.shape, df_test.shape
```

*Las dimensiones (número de filas y columnas) de df_train son 310 FILAS y 17 COLUMNAS y del df_test son 162 FILAS y 17 columnas*

```python
# Defino como hiperparametro windows_size con un valor de 7. El tamaño de la ventana determinará cuántas observaciones pasadas se utilizarán para predecir el valor actual.
# Como el dataset es relativamente pequeño (472 filas) es recomendable utilizar un valor de window_size pequeño,
# ya que un window_size más grande puede llevar a la pérdida de información y a un sobreajuste en los datos de entrenamiento.

window_size = 9
```

```python
# Caracterizo los conjunto de entrenamiento y de prueba, para cada una de las columnas, creando "columnas desplazadas" (para esto es útil la función shift) y concatenandolas (pd.concat).

cols = df.columns # Asigno todas las columnas del DataFrame a la variable cols.

X_train = pd.concat([pd.concat([df_train[c].shift(w, fill_value=0) for w in range(1, window_size+1)], axis=1, keys=[f"{c}_{w}" for w in range(1, window_size+1)]) for c in cols], axis=1)
X_test  = pd.concat([pd.concat([df_test[c].shift(w, fill_value=0) for w in range(1, window_size+1)], axis=1, keys=[f"{c}_{w}" for w in range(1, window_size+1)]) for c in cols], axis=1)
```

```python
# Señalo que se deberá predecir la columna "greatest losses direction" (dirección de las mayores pérdidas).

y_train = df_train["greatest losses direction"]
y_test = df_test["greatest losses direction"]

X_train.head(), y_train.head()
```

```python
# Preparo los conjuntos de datos de entrenamiento y prueba en el formato adecuado para ser procesados por el modelo LSTM.
# LSTM espera una entrada en forma de tensor tridimensional

X_train = X_train.values.reshape( (X_train.shape[0], 1, X_train.shape[1]) ) # Mantengo todas las filas (X_train.shape[0]), una dimensión adicional de tamaño 1, y mantengo el nro. original de características (X_train.shape[1])
X_test = X_test.values.reshape( (X_test.shape[0], 1, X_test.shape[1]) ) # Mismas caracteristicas que X_train
```

```python
X_train.shape
```

```python
X_test.shape
```

*(número de ejemplos, número de pasos de tiempo, número de características)*

```python
# Modelo Keras: LSTM multivariado

model = keras.models.Sequential([
    keras.layers.InputLayer(input_shape=[X_train.shape[1], X_train.shape[2]]),

    keras.layers.LSTM(units=24, return_sequences=True), # return_sequences=True para pasar la secuencia de salida a la siguiente capa LSTM || 24 unidades de LSTM en esta capa
    keras.layers.LSTM(units=12), # 12 unidades LSTM en esta capa

    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(1, activation="linear")
])

model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mean_squared_error"])
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=5, batch_size=50, verbose=True)
```

```python
# Predicciones del modelo
predictions = model.predict(X_test)
```

```python
# Evaluo el rendimiento del modelo con RMSE (raíz cuadrada de la media de los errores cuadráticos entre las predicciones y los valores reales)
## (!) Cuanto menor sea el valor del RMSE, mejor será el rendimiento del modelo en términos de la capacidad de ajuste y la precisión de las predicciones

np.sqrt(np.mean((predictions.flatten() - y_test)**2))
```

Las predicciones del modelo se desvían en aproximadamente 0.5212 unidades
